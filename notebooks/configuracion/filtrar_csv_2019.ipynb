```python
import pandas as pd
import glob
import os

# Ruta a la carpeta con los archivos CSV
ruta_carpeta = "/home/sfp/Documents/Drive/UPV/Segundo/2ºCuatri/Proyecto/estaciones_merged"

# Obtener la lista de archivos CSV en la carpeta
archivos = [f for f in glob.glob(os.path.join(ruta_carpeta, "*.csv")) if os.path.isfile(f)]

# Año mínimo de interés
anio_minimo = 2019

# Procesar cada archivo
for archivo in archivos:
    try:
        # Leer el archivo CSV
        df = pd.read_csv(archivo)
        
        # Convertir la columna de fecha a formato datetime (asegurando que el día sea primero)
        df["FECHA"] = pd.to_datetime(df["FECHA"], format="%d/%m/%Y", errors="coerce")
        
        # Eliminar filas donde la conversión falló (valores NaT)
        df = df.dropna(subset=["FECHA"])
        
        # Filtrar solo los datos de 2019 en adelante
        df_filtrado = df[df["FECHA"].dt.year >= anio_minimo]
        
        # Obtener el nombre del archivo original
        nombre_archivo = os.path.basename(archivo)
        
        # Crear un nuevo nombre de archivo (ejemplo: "filtrado_nombre.csv")
        nuevo_nombre = f"filtrado_{nombre_archivo}"
        ruta_salida = os.path.join(ruta_carpeta, nuevo_nombre)
        
        # Guardar el archivo filtrado
        df_filtrado.to_csv(ruta_salida, index=False)
        
        print(f"✅ Archivo guardado: {nuevo_nombre} ({len(df_filtrado)} filas)")

    except Exception as e:
        print(f"❌ Error procesando {archivo}: {e}")
```
