```python
import pandas as pd

df = pd.read_parquet("../../data/cleaned/calidad_aire/base_para_obj1.parquet")

mask = df['FechaHora'].str.len() == 10

df.loc[mask, 'FechaHora'] = (
    df.loc[mask, 'FechaHora']
    + ' '
    + df.loc[mask, 'Hora'].astype(str).str.zfill(2)
    + ':00:00'
)

df['FechaHora'] = pd.to_datetime(
    df['FechaHora'], format='%Y-%m-%d %H:%M:%S'
)

df.to_parquet(
    "../../data/cleaned/calidad_aire/base_para_obj1.parquet",
    index=False,
    engine='pyarrow'
)
```
