```python
df = pd.read_csv('../../data/raw/calidad_aire/conjunto_estaciones.csv')
```

```python
# …existing code…

# 1) Intentar inferir numéricos/fechas en columnas object
for col in df.select_dtypes(include=['object']).columns:
    # a) ¿puede ser numérico?
    tmp = pd.to_numeric(df[col], errors='coerce')
    if tmp.notna().sum() > 0:
        df[col] = tmp
        continue
    # b) ¿puede ser fecha?
    tmp = pd.to_datetime(df[col], errors='coerce', dayfirst=False)
    if tmp.notna().sum() > 0:
        df[col] = tmp
        continue
    # c) si no, asegurar que es string puro
    df[col] = df[col].astype(str)

```

```python
df.to_parquet('../../data/raw/calidad_aire/conjunto_estaciones.parquet')
```
