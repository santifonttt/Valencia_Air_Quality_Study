```python
import pandas as pd
```

```python
estaciones = pd.read_csv('../../data/cleaned/calidad_aire/conjunto_estaciones_limpio.csv')
```

```python
estaciones.info()
```

```python
estaciones.head()
```

```python
float_cols = estaciones.select_dtypes(include=['float64']).columns
estaciones[float_cols] = estaciones[float_cols].astype('float32')
```

```python
int_cols = estaciones.select_dtypes(include=['int64']).columns
int_cols
```

```python
estaciones[["Hora", "Dia", "MES_NUMERO"]] = estaciones[["Hora", "Dia", "MES_NUMERO"]].astype("int8")
estaciones["Año"] = estaciones["Año"].astype("int16")
```

```python
estaciones.info()
```

```python
estaciones.to_feather('../../data/cleaned/calidad_aire/conjunto_estaciones_limpio.feather')
```
