```python
import os
import pandas as pd

root_folder = '/home/sfp/Documents/Drive/UPV/Segundo/2ºCuatri/Proyecto/Analisis_aire/data/Calidad Aire/Por Zona'

def merge_csv_by_folder(root_folder):
    # Recorre todas las carpetas dentro de 'Por Zona'
    for town_folder in os.listdir(root_folder):
        town_path = os.path.join(root_folder, town_folder)
        
        # Verifica que sea un directorio
        if os.path.isdir(town_path):
            csv_files = [f for f in os.listdir(town_path) if f.lower().endswith('.csv')]
            print(f'Procesando carpeta: {town_folder}, Archivos encontrados: {csv_files}')
            
            # Lista para almacenar los DataFrames
            dataframes = []
            
            for csv_file in csv_files:
                csv_path = os.path.join(town_path, csv_file)
                try:
                    df = pd.read_csv(csv_path, encoding='utf-8')  # Intentar con UTF-8
                    dataframes.append(df)
                except UnicodeDecodeError:
                    print(f'Error de codificación en {csv_path}, intentando con ISO-8859-1')
                    try:
                        df = pd.read_csv(csv_path, encoding='ISO-8859-1')  # Intentar con otra codificación
                        dataframes.append(df)
                    except Exception as e:
                        print(f'Error leyendo {csv_path}: {e}')
            
            # Concatenar todos los DataFrames en uno solo
            if dataframes:
                merged_df = pd.concat(dataframes, ignore_index=True)
                output_file = os.path.join(town_path, f'{town_folder}_merged.csv')
                merged_df.to_csv(output_file, index=False, encoding='utf-8')
                print(f'Merged CSV saved: {output_file}')
            else:
                print(f'No se encontraron archivos CSV en {town_folder} o hubo errores.')

# Ruta de la carpeta principal
root_folder = "/home/sfp/Documents/Drive/UPV/Segundo/2ºCuatri/Proyecto/Analisis_aire/data/Calidad Aire/Por Zona"
merge_csv_by_folder(root_folder)
```

```python
import os
import pandas as pd

# Ruta de la carpeta donde están los archivos CSV
folder_path = '/home/sfp/Documents/Drive/UPV/Segundo/2ºCuatri/Proyecto/estaciones_merged/filtrados/final'

def merge_csv_in_folder(folder_path):
    # Lista para almacenar los DataFrames
    dataframes = []
    
    # Recorre todos los archivos en la carpeta
    csv_files = [f for f in os.listdir(folder_path) if f.lower().endswith('.csv')]
    print(f'Archivos CSV encontrados: {csv_files}')
    
    for csv_file in csv_files:
        csv_path = os.path.join(folder_path, csv_file)
        try:
            # Intentar leer el archivo CSV
            df = pd.read_csv(csv_path, encoding='utf-8')  # Intentar con UTF-8
            df['archivo_origen'] = csv_file  # Agregar columna con el nombre del archivo
            dataframes.append(df)
        except UnicodeDecodeError:
            print(f'Error de codificación en {csv_path}, intentando con ISO-8859-1')
            try:
                df = pd.read_csv(csv_path, encoding='ISO-8859-1')  # Intentar con ISO-8859-1
                df['archivo_origen'] = csv_file  # Agregar columna con el nombre del archivo
                dataframes.append(df)
            except Exception as e:
                print(f'Error leyendo {csv_path}: {e}')
    
    # Concatenar todos los DataFrames en uno solo, alineando las columnas
    if dataframes:
        merged_df = pd.concat(dataframes, ignore_index=True, sort=False)
        output_file = os.path.join(folder_path, 'datos_final.csv')  # Guardar el archivo final en la misma carpeta
        merged_df.to_csv(output_file, index=False, encoding='utf-8')
        print(f'Merged CSV saved: {output_file}')
    else:
        print('No se encontraron archivos CSV o hubo errores en la lectura.')

# Llamar a la función para combinar los CSVs
merge_csv_in_folder(folder_path)

```
